```python
import graphlab
```

```python
products = graphlab.SFrame.read_csv('C:/Users/dcvr1/OneDrive - IIIT Bangalore/7th semester/ML/Project/Reviews.csv')
```

```python
products = products.select_columns(['Score','Text','HelpfulnessDenominator','HelpfulnessNumerator'])
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['Text'])
```

```python
products.head()
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
products = products[products['Score'] != 3]
```

```python
len(products)
```

```python
products['sentiment'] = products['Score'] >=4
```

```python
products.head()
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
len(train_data)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,target='sentiment',features=['word_count'],
validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```

```python
products['predictions'] = sentiment_model.predict(products, output_type='probability')
```

```python
products.head()
```

```python
sorted_reviews = products.sort('predictions', ascending=False)
```

```python
products["Usefulness"] = (products["HelpfulnessNumerator"]/products["HelpfulnessDenominator"]).apply(lambda n: "1" if n > 0.8 else "0")
```

```python
products.head()
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
useful_model = graphlab.logistic_classifier.create(train_data,target='Usefulness',features=['word_count'],
validation_set=test_data)
```

```python
useful_model.evaluate(test_data, metric='roc_curve')
```
